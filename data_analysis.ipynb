```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib qt
```

```python
# TRIANGULAR
FD_T4 = np.loadtxt('FD_T4.txt')
FD_T6 = np.loadtxt('FD_T6.txt')
FD_T8 = np.loadtxt('FD_T8.txt')


# QUADRADO
FD_S4 = np.loadtxt('FD_S4.txt')
FD_S6 = np.loadtxt('FD_S6.txt')
FD_S8 = np.loadtxt('FD_S8.txt')

# CIRCULO 
FD_C4 = np.loadtxt('FD_C4.txt')
FD_C6 = np.loadtxt('FD_C6.txt')
FD_C8 = np.loadtxt('FD_C8.txt')
```

```python
font = {'font':'serif'}
plt.plot(FD_T4, label='j = 4', color='k')
plt.plot(FD_T6, label='j = 6', color='blue')
plt.plot(FD_T8, label='j = 8', color='red')
plt.axhline(2.0, color='k', ls='--')
plt.title("Gro Triangular", fontdict=font, size = 20)
plt.xlabel('Passos', fontdict=font, size=15)
plt.ylabel('Dimenso Fractal', fontdict=font, size= 15)
plt.grid()
plt.legend(title='Simetrias', shadow=True, fancybox=True, facecolor='lightgray', title_fontsize='large')
plt.tight_layout()
```

```python
font = {'font':'serif'}
plt.plot(FD_S4, label='j = 4', color='k')
plt.plot(FD_S6, label='j = 6', color='blue')
plt.plot(FD_S8, label='j = 8', color='red')
plt.axhline(2.0, color='k', ls='--')
plt.title("Gro Quadrado", fontdict=font, size = 20)
plt.xlabel('Passos', fontdict=font, size=15)
plt.ylabel('Dimenso Fractal', fontdict=font, size= 15)
plt.grid()
plt.legend(title='Simetrias', shadow=True, fancybox=True, facecolor='lightgray', title_fontsize='large')
plt.tight_layout()
```

```python
font = {'font':'serif'}
plt.plot(FD_C4, label='j = 4', color='k')
plt.plot(FD_C6, label='j = 6', color='blue')
plt.plot(FD_C8, label='j = 8', color='red')
plt.axhline(2.0, color='k', ls='--')
plt.title("Gro Circular", fontdict=font, size = 20)
plt.xlabel('Passos', fontdict=font, size=15)
plt.ylabel('Dimenso Fractal', fontdict=font, size= 15)
plt.grid()
plt.legend(title='Simetrias', shadow=True, fancybox=True, facecolor='lightgray', title_fontsize='large')
plt.tight_layout()
```

Plot Geral das Dimenses Fractais

```python
plt.style.use("seaborn-v0_8-poster")
figure, axs = plt.subplots(ncols=3, nrows=1, figsize=(20,6)) 

axs[0].set_title("Tringulo"); 
axs[0].plot(FD_T4[:], label='j=4')
axs[0].plot(FD_T6[:], label='j=6')
axs[0].plot(FD_T8[:], label='j=8')
axs[0].axvline(100, ls='--', c='k')
axs[0].legend()

axs[1].set_title("Quadrado"); 
axs[1].plot(FD_S4[:], label='j=4')
axs[1].plot(FD_S6[:], label='j=6')
axs[1].plot(FD_S8[:], label='j=8')
axs[1].axvline(100, ls='--', c='k')

# axs[1].set_xlim([100, 729])
axs[1].legend()

axs[2].set_title("Crculo"); 
axs[2].plot(FD_C4[:], label='j=4')
axs[2].plot(FD_C6[:], label='j=6')
axs[2].plot(FD_C8[:], label='j=8')
axs[2].axvline(100, ls='--', c='k')

axs[2].legend()
```

```python

font = {'family':'serif'}
# Create the main figure and axes
fig, ax_main = plt.subplots()

# Plot the main graph
ax_main.set_title("Semente Triangular", fontdict=font, size=35, fontweight='bold')
ax_main.tick_params(axis='both', which='major', labelsize=30)
ax_main.plot(FD_T4, label='j=4', color = 'k')
ax_main.plot(FD_T6, label='j=6', color = 'blue')
ax_main.plot(FD_T8, label='j=8',color = 'red')


# Create the inset axes
left, bottom, width, height = 0.71, 0.52, 0.25, 0.25
ax_inset = fig.add_axes([left, bottom, width, height])

# Plot the sub-graph (zoomed in version)
ax_inset.plot(FD_T4[-100:], color='k')
ax_inset.plot(FD_T6[-100:], color='blue')
ax_inset.plot(FD_T8[-100:], color='red')

ax_inset.set_title("ltimos 100 passos")

# Add labels and legends
ax_main.set_xlabel('Passos', fontdict=font, size=30 )
ax_main.set_ylabel('Dim. Fractal', fontdict=font, size=30)
ax_main.grid()
ax_main.legend(title='Simetrias', shadow=True, fancybox=True, facecolor='lightgray', title_fontsize='large', fontsize=25)

# ax_inset.set_xlabel('X-axis (Zoomed)')
# ax_inset.set_ylabel('Y-axis (Zoomed)')
ax_inset.legend()
plt.tight_layout()
plt.show()
```
